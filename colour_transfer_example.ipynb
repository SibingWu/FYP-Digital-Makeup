```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import logging
```

```python
logging.basicConfig(level=logging.INFO)
```

```python
from colour_transfer import colour_transfer_by_pdf_transfer
```

```python
# read in images
img_original = plt.imread('scotland_house.png')
img_target = plt.imread('scotland_plain.png')

df_original = pd.DataFrame(img_original.reshape(-1, img_original.shape[-1]), columns=['r', 'g', 'b'])
df_target = pd.DataFrame(img_target.reshape(-1, img_target.shape[-1]), columns=['r', 'g', 'b'])

arr_original = df_original.values
arr_target = df_target.values
```

```python
df_original.head()
```

```python
arr_original
```

```python
# apply colour transfer
df_result = pd.DataFrame(colour_transfer_by_pdf_transfer(arr_original, arr_target), columns=['r', 'g', 'b'])

arr_result = df_result.values
img_result = arr_result.reshape(img_original.shape)
```

```python
# visualize the result
fig, ax = plt.subplots(1, 3, figsize=(18,6))
ax[0].imshow(img_original)
ax[0].set_title('original')
ax[1].imshow(img_target)
ax[1].set_title('target')
ax[2].imshow(img_result)
ax[2].set_title('result')
```
